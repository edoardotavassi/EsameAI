```python
import treasure_maze as tm
import maze_generator as mg
```

```python
arrAstar=list()
arrDfs=list()
arrBfs=list()
timeAstar=0
iterationsAstar=0
costAstar=0

timeDfs=0
iterationsDfs=0
costDfs=0

timeBfs=0
iterationsBfs=0
costBfs=0
repetitions=100
for i in range(2, 5):
    for k in range (0,repetitions):
        print("Dimensione labirinto: ", i)
        print("Iterazione: ", k)
        maze = mg.Maze(i, i)
        map = maze.testGeneration()
        print(map)
        problem = tm.TreasureMaze(map)
        (time, iterations, node)= tm.astar_search_graph(problem)
        timeAstar += time
        iterationsAstar += iterations
        costAstar += node.path_cost
        
        problem = tm.TreasureMaze(map)
        (time, iterations, node)= tm.depth_first_graph_search(problem)
        timeDfs += time
        iterationsDfs += iterations
        costDfs += node.path_cost
        
        problem = tm.TreasureMaze(map)
        (time, iterations, node)= tm.breadth_first_graph_search(problem)
        timeBfs += time
        iterationsBfs += iterations
        costBfs += node.path_cost
        
    
    arrAstar.append((i,timeAstar/repetitions, iterationsAstar/repetitions, costAstar/repetitions))
    arrDfs.append((i,timeDfs/repetitions, iterationsDfs/repetitions, costDfs/repetitions))
    arrBfs.append((i,timeBfs/repetitions, iterationsBfs/repetitions, costBfs/repetitions))
```

```python
print (arrAstar)
print(arrDfs)
print(arrBfs)
```

```python
import matplotlib.pyplot as plt
```

```python
outAstarTime=list()
outDfsTime=list()
outBfsTime=list()
for i in range(0,3):
    outAstarTime.append((arrAstar[i][0], arrAstar[i][1]))
    
for i in range(0,3):
    outDfsTime.append((arrDfs[i][0], arrDfs[i][1]))
    
for i in range(0,3):
    outBfsTime.append((arrBfs[i][0], arrBfs[i][1]))

zip(*outAstarTime)
zip(*outDfsTime)
zip(*outBfsTime)
   
plt.plot(*zip(*outAstarTime), label="A*")
plt.plot(*zip(*outDfsTime), label="DFS")
plt.plot(*zip(*outBfsTime), label="BFS")
#plt.plot(outDfsTime)
#plt.plot(outBfsTime)
plt.ylim(ymax = 0.08, ymin = 0)
plt.xlim(xmax = 4, xmin = 2)
plt.title('Tempo')
plt.ylabel('Time')
plt.xlabel('Matrix size')
plt.legend(['A*', 'DFS', "BFS"], loc='upper left')
plt.show()
```
